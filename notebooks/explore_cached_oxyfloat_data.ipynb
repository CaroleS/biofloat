### Explore locally cached Argo oxygen float data
*Use the oxyfloat module to get data and Pandas to operate on it*

Before executing this notebook a local HDF file cache needs to populated by executing this script:

```bash
scripts/load_cache.py --age 1000 --max_profiles 3 -v
```
Then we can use oxyfloat and this Notebook to interactively explore the data.

````python
from oxyfloat import OxyFloat
of = OxyFloat(cache_file='/home/mccann/dev/oxyfloatgit/oxyfloat/oxyfloat_age_1000_max_profiles_3.hdf')
````

````python
%%time
wmo_list = of.get_oxy_floats_from_status(age_gte=1000)
df = of.get_float_dataframe(wmo_list, max_profiles=1)
````

````python
%pylab inline
import pylab as plt
from mpl_toolkits.basemap import Basemap

plt.rcParams['figure.figsize'] = (18.0, 8.0)
m = Basemap(llcrnrlon=15, llcrnrlat=-90, urcrnrlon=390, urcrnrlat=90, projection='cyl')
m.fillcontinents(color='0.8')

m.scatter(df['LONGITUDE'], df['LATITUDE'], latlon=True)
````
